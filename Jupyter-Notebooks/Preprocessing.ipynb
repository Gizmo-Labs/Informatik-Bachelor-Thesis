```python
import pandas as pd # type: ignore
import numpy as np # type: ignore
```

#### ▶️ Einlesen der aufgezeichneten Daten von Subjekt 1 aus dem Quellordner

- Testperson trug das Myo-Armband am rechten Arm

```python
# Importiere die .csv-Dateien mit den EMG-Daten der einzelnen Subjekte und den einzelnen Labels (Gesten)
# Dabei gilt folgendes Schema, siehe auch Bilder im Ordner "Gestenbilder"
# Label 0 --> Papier
# Label 1 --> Stein
# Label 2 --> Schere


##############################################################
# Aufnahme der Einzelgesten von Subjekt 1                    #
# Datentyp: Python List                                      #
##############################################################
subject1_paper = []
subject1_rock = []
subject1_scissors = []

##############################################################
# Erstelle Dateipfad und lese Datei ein                      #
# Es existieren 6 x 10 Wiederholungen pro Geste              #
##############################################################
for session in range(1,7):
    for attempt in range(0, 9):
        subject1_paper.append(pd.read_csv('./Datasets/Proband1/Separated/s1_r_' + str(session) + '-paper-' + str(attempt) + '-emg.csv', sep=',', header='infer').iloc[:, 2:10])        
        subject1_rock.append(pd.read_csv('./Datasets/Proband1/Separated/s1_r_' + str(session) + '-rock-' + str(attempt) + '-emg.csv', sep=',', header='infer').iloc[:, 2:10])        
        subject1_scissors.append(pd.read_csv('./Datasets/Proband1/Separated/s1_r_' + str(session) + '-scissors-' + str(attempt) + '-emg.csv', sep=',', header='infer').iloc[:, 2:10])        
    

##############################################################
# Schreibe alle Sub-Dataframes in ein gesamtes Dataframe     #
# Datentyp: Pandas Dataframe                                 #
##############################################################
subject1_paper = pd.concat([df for df in subject1_paper], axis=0).reset_index(drop=True)
subject1_rock = pd.concat([df for df in subject1_rock], axis=0).reset_index(drop=True)
subject1_scissors = pd.concat([df for df in subject1_scissors], axis=0).reset_index(drop=True)

subject1_paper.iloc[0:400].plot(title='Papier',legend=False)
subject1_rock.iloc[0:400].plot(title='Stein',legend=False)
subject1_scissors.iloc[0:400].plot(title='Schere',legend=False)


##############################################################
# Schreibe alle Daten in eine .csv-Datei für jede Geste      #
# Datentyp: Pandas Dataframe                                 #
##############################################################
subject1_paper.to_csv('./Datasets/Concatenated/S1L0.csv', index=False, header=None)
subject1_rock.to_csv('./Datasets/Concatenated/S1L1.csv', index=False, header=None)
subject1_scissors.to_csv('./Datasets/Concatenated/S1L2.csv', index=False, header=None)
```

```python
# Importiere die .csv-Dateien mit den EMG-Daten der einzelnen Subjekte und den einzelnen Labels (Gesten)
# Dabei gilt folgendes Schema, siehe auch Bilder im Ordner "Gestenbilder"
# Label 0 --> Papier
# Label 1 --> Stein
# Label 2 --> Schere


##############################################################
# Aufnahme der Einzelgesten von Subjekt 9                    #
# Datentyp: Python List                                      #
##############################################################
subject10_paper = []
subject10_rock = []
subject10_scissors = []

##############################################################
# Erstelle Dateipfad und lese Datei ein                      #
# Es existieren 6 x 10 Wiederholungen pro Geste              #
##############################################################
for session in range(1,4):
    for attempt in range(0, 9):
        subject10_paper.append(pd.read_csv('./Datasets/Proband10/Separated/s10_r_' + str(session) + '-paper-' + str(attempt) + '-emg.csv', sep=',', header='infer').iloc[:, 2:10])        
        subject10_rock.append(pd.read_csv('./Datasets/Proband10/Separated/s10_r_' + str(session) + '-rock-' + str(attempt) + '-emg.csv', sep=',', header='infer').iloc[:, 2:10])        
        subject10_scissors.append(pd.read_csv('./Datasets/Proband10/Separated/s10_r_' + str(session) + '-scissors-' + str(attempt) + '-emg.csv', sep=',', header='infer').iloc[:, 2:10])        
    

##############################################################
# Schreibe alle Sub-Dataframes in ein gesamtes Dataframe     #
# Datentyp: Pandas Dataframe                                 #
##############################################################
subject10_paper = pd.concat([df for df in subject10_paper], axis=0).reset_index(drop=True)
subject10_rock = pd.concat([df for df in subject10_rock], axis=0).reset_index(drop=True)
subject10_scissors = pd.concat([df for df in subject10_scissors], axis=0).reset_index(drop=True)

subject10_paper.iloc[500:800].plot(title='Papier',legend=False)
subject10_rock.iloc[500:800].plot(title='Stein',legend=False)
subject10_scissors.iloc[500:800].plot(title='Schere',legend=False)


##############################################################
# Schreibe alle Daten in eine .csv-Datei für jede Geste      #
# Datentyp: Pandas Dataframe                                 #
##############################################################
subject10_paper.to_csv('./Datasets/Concatenated/S10L0.csv', index=False, header=None)
subject10_rock.to_csv('./Datasets/Concatenated/S10L1.csv', index=False, header=None)
subject10_scissors.to_csv('./Datasets/Concatenated/S10L2.csv', index=False, header=None)
```
